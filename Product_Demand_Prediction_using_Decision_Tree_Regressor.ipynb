```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv('/content/Product_Demand_Prediction_Dataset.csv')
data.head()
```

```python
data = data.dropna()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Total Price', y='Units Sold', size='Units Sold',color='green')
plt.show()
```

```python
X = data[["Total Price", "Base Price"]]
Y = data["Units Sold"]
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,test_size=0.2, random_state=42)
model = DecisionTreeRegressor()
model.fit(Xtrain,Ytrain)
```

```python
features = pd.DataFrame({'Total Price': [133.04], 'Base Price': [140.89]})
Prediction = model.predict(features)
print(Prediction)
```
